```
# To create inteactive candlestick chart in Python using Plotly
```

```
# Import libraries
import pandas as pd
import plotly.graph_objects as go
import pandas_datareader as web
```

```
df=web.DataReader('RELIANCE.NS',data_source="yahoo",start='2020-01-01',end='2020-07-24')
df
```

```
#create candlestick chart
figure = go.Figure(
    data = go.Candlestick(
        x=df.index,
        low=df["Low"],
        high=df["High"],
        open=df["Open"],
        close=df["Close"]
    )
)
# to remove the slider
#figure.update_layout(xaxis_rangeslider_visible = False)
#to vary the height
figure.update_layout(height = 700)

figure.update_layout(
    title = "Reliance Stock Price",
    yaxis_title = "Price in $",
    xaxis_title = "Date"
)

figure.show()
```
